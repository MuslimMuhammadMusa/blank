```python
import random
import binarytree
from graphviz import Digraph
import turtle

# Create a list of 15 random integers
lst = random.sample(range(1, 101), 15)

# Define a Node class for binary tree
class Node:
    def __init__(self, value=None):
        self.value = value
        self.left_child = None
        self.right_child = None

# Create a function to build the binary tree
def build_tree(lst):
    root = Node(lst[0])
    for i in range(1, len(lst)):
        node = Node(lst[i])
        curr = root
        while True:
            if node.value < curr.value:
                if curr.left_child is None:
                    curr.left_child = node
                    break
                else:
                    curr = curr.left_child
            else:
                if curr.right_child is None:
                    curr.right_child = node
                    break
                else:
                    curr = curr.right_child
    return root

# Build the binary tree and print it using binarytree library
root = build_tree(lst)
print(binarytree.build_tree(lst))

# Visualize the binary tree using graphviz
def visualize_tree(root):
    dot = Digraph()
    def add_edges(node):
        if node.left_child:
            dot.edge(str(node.value), str(node.left_child.value))
            add_edges(node.left_child)
        if node.right_child:
            dot.edge(str(node.value), str(node.right_child.value))
            add_edges(node.right_child)
    add_edges(root)
    dot.render('tree.gv', view=True)

visualize_tree(root)

# Visualize the binary tree using turtle graphics
def turtle_tree(root, x, y, x_dist, y_dist, t):
    if root is not None:
        t.penup()
        t.goto(x, y)
        t.pendown()
        t.write(root.value, align='center', font=('Arial', 12, 'normal'))
        turtle_tree(root.left_child, x-x_dist, y-y_dist, x_dist/2, y_dist-20, t)
        turtle_tree(root.right_child, x+x_dist, y-y_dist, x_dist/2, y_dist-20, t)

t = turtle.Turtle()
t.speed(0)
turtle.bgcolor('white')
t.hideturtle()
turtle_tree(root, 0, 300, 400, 50, t)
turtle.done()
```
