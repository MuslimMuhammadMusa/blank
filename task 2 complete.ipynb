```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def build_bst(numbers):
    # Sort the list of numbers
    numbers.sort()

    # Define a recursive function to build the binary search tree
    def build_tree(start, end):
        if start > end:
            return None

        # Find the middle element and make it the root of the subtree
        mid = (start + end) // 2
        root = Node(numbers[mid])

        # Recursively build the left and right subtrees
        root.left = build_tree(start, mid - 1)
        root.right = build_tree(mid + 1, end)

        return root

    # Call the recursive function to build the binary search tree
    root = build_tree(0, len(numbers) - 1)

    return root
```

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def bfs(root):
    if root is None:
        return []

    # Create a queue and add the root node to it
    queue = deque()
    queue.append(root)

    # Process each node in the queue
    result = []
    while len(queue) > 0:
        # Get the next node from the queue
        node = queue.popleft()

        # Add the node value to the result
        result.append(node.value)

        # Add the left and right children of the node to the queue
        if node.left is not None:
            queue.append(node.left)
        if node.right is not None:
            queue.append(node.right)

    return result
```

```python
import pandas as pd
import time
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def build_bst(numbers):
    if not numbers:
        return None
    mid = len(numbers) // 2
    root = Node(numbers[mid])
    root.left = build_bst(numbers[:mid])
    root.right = build_bst(numbers[mid+1:])
    return root

def bfs(root, goal):
    queue = deque([(root, 0)])
    while queue:
        node, level = queue.popleft()
        if node.value == goal:
            return level
        if node.left:
            queue.append((node.left, level+1))
        if node.right:
            queue.append((node.right, level+1))
    return None

def dfs(node, goal, level):
    if node is None:
        return None
    if node.value == goal:
        return level
    left_result = dfs(node.left, goal, level+1)
    if left_result is not None:
        return left_result
    right_result = dfs(node.right, goal, level+1)
    if right_result is not None:
        return right_result
    return None

# Generate shuffled list of numbers
numbers = list(range(1001))
random.shuffle(numbers)

# Build the binary tree
root = build_bst(numbers)

# Run BFS and record time
goal = numbers[len(numbers)-220]
start_time_bfs = time.time()
bfs_result = bfs(root, goal)
end_time_bfs = time.time()
time_taken_bfs = end_time_bfs - start_time_bfs

# Run DFS and record time
start_time_dfs = time.time()
dfs_result = dfs(root, goal, 0)
end_time_dfs = time.time()
time_taken_dfs = end_time_dfs - start_time_dfs

# Create DataFrame
df = pd.DataFrame({
    'Algorithm': ['BFS', 'DFS'],
    'Time Taken': [time_taken_bfs, time_taken_dfs],
    'Result': [bfs_result, dfs_result]
})

# Print the DataFrame
print(df)
```

```python
import random
import time
import pandas as pd

# Function to generate a list of unique random numbers
def generate_random_numbers(n):
    return random.sample(range(1, n+1), n)

# Function to build a binary tree from a list of numbers
def build_binary_tree(nums):
    class Node:
        def __init__(self, val):
            self.val = val
            self.left = None
            self.right = None
            
    root = Node(nums[0])
    for num in nums[1:]:
        curr = root
        while True:
            if num < curr.val:
                if curr.left is None:
                    curr.left = Node(num)
                    break
                else:
                    curr = curr.left
            else:
                if curr.right is None:
                    curr.right = Node(num)
                    break
                else:
                    curr = curr.right
    return root

# Function to perform BFS on a binary tree and find the goal
def bfs(root, goal):
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node.val == goal:
            return True
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return False

# Function to perform DFS on a binary tree and find the goal
def dfs(root, goal):
    stack = [root]
    while stack:
        node = stack.pop()
        if node.val == goal:
            return True
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return False

# List of tree sizes
sizes = [1000, 40000, 80000, 200000, 1000000]

# List of results
results = []

# Loop through each tree size
for size in sizes:
    # Generate random and unique numbers for the tree
    nums = generate_random_numbers(size)
    # Build the binary tree
    root = build_binary_tree(nums)
    # Set the goal
    goal = nums[len(nums) - 220]
    # Perform BFS and measure the time taken
    start_time = time.time()
    bfs_found = bfs(root, goal)
    bfs_time = time.time() - start_time
    # Perform DFS and measure the time taken
    start_time = time.time()
    dfs_found = dfs(root, goal)
    dfs_time = time.time() - start_time
    # Append the results to the list
    results.append({'Tree Size': size, 'BFS Time': bfs_time, 'DFS Time': dfs_time})

# Create a data frame from the results
df = pd.DataFrame(results)

# Plot the bar chart
import matplotlib.pyplot as plt

plt.bar(df['Tree Size'], df['BFS Time'], width=5000, align='center', alpha=0.5)
plt.bar(df['Tree Size'], df['DFS Time'], width=5000, align='center', alpha=0.5)
plt.legend(['BFS Time', 'DFS Time'])
plt.xlabel('Tree Size')
plt.ylabel('Time (seconds)')
plt.title('BFS vs. DFS Search Time for Different Tree Sizes')
plt.xticks(df['Tree Size'])
plt.show()
```
